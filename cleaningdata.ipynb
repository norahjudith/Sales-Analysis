```python
import pandas as pd
data = pd.read_csv("../input/masterclass/Walmart.csv")
data
```

```python
data.dropna(how = "all", inplace = True)
data
```

```python
data["Country"].replace({"United States":"USA"}, inplace = True)
data
```

```python
avg=data["Sales"].mean()
data["Sales"].fillna(value = avg, inplace = True)
data
```

```python
data.to_csv("CleanedData.csv")
```
